```python
# These are standard Python libraries when doing machine learning analysis. 
import pandas as pd
import numpy as np 

# These libraries are used for plots and visualizations. 
import matplotlib.pyplot as plt
import seaborn as sns
```

Our dataset came from <a href="https://www.kaggle.com/datasets/febinphilips/us-house-listings-2023/data"> Kaggle </a> , sourced from Zillow, a leading online platform for real estate transactions giving information on approximately 100 million homes. This dataset offers valuable perspectives on real estate patterns and property attributes. Download the the dataset from Kaggle to be able to import it. 


```python
# This is how you import the CSV file. 
df = pd.read_csv("zillow_housing_prices.csv")

print(df.head())

row_count = len(df)

# Printing the number of rows 

print("\n")
print(f'The Dataset has {row_count} observations.')
```

Our different features in the data set are:

<h2 id="Data-Processing">Part 2: Data Processing/Cleaning<a class="anchor-link" href="#Data-Processing"></a></h2>

Now that we imported our data. First we're going to clean our data to be usable and make sure it's relevant to our mission. 

Our goal is to predict the predict the house market estimates. Thus, we will drop city and street because they are too specific for our purposes. We only want an overall estimate of house prices per state. 

Out of Area, PPSq, LotArea, Lot Unit, and Converted Lot, we will be dropping Lot Area, Lot Unit and Converted because that gives us the total land area in acres. Since we already have area in square feet, we do not need any of those columns to keep it consistent. 

Out of Listed Price, Market Estimate, and Rent Estimate,  we are dropping Listed Price because that's the last listed price of the property. We are only interested in predicting the market estimate of houses so we will drop Rent Estimate as well. 

We will also drop Latitude and Longitude because those aren't relevant to estimating the market price if we keep the states. 

```python
df.drop(columns=['City', 'Street', 'Zipcode', 'Latitude','Longitude', 'LotArea', 'LotUnit', 'ConvertedLot', 'RentEstimate', 'Price'], inplace = True)
print(df.columns)
```

The next concern we have about cleaning our data is missing data. We have some missing values like NAN's in our Market Estimate. Our other columns also have missing data. 

First we'll classify our missing data to the three categories of Missing Completely at Random (MCAR), Missing at Random (MAR), and Missing Not at Random (MNAR) scenarios. More information can be read <a href ="https://www.ncbi.nlm.nih.gov/books/NBK493614/"> here </a>. 

Some of the data in the columns such as Bedroom, Bathroom, and Area have missing values for features due to rare cases or listings not providing the data. This can be considered Missing Completely at Random (MCAR), because the data isn't related to the other observed or unobserved variables int he dcataset. 

The missingingness in "MarketEstimate" can be considered Missing at Random (MAR) if it's related to other observed variables in the dataset. For instance, the missingingness could be due to Zillow choosing not to provide estimates based on certain characteristics of the properties (like location or size), then the missingness of these values is related to observed variables in the dataset.

There are several ways you can approach missing data. We could doing missing point imputation which replace missing data with substitute values based on its closest neighbors. We could also deal with missing data by just dropping the relevant rows. 

Looking at our missing data, since we have several rows of data (24521 rows), we can afford to just drop the rows of missing data. We don't want to overfit our model later on the data. 

```python
# Dropping all rows with missing data 
df.dropna(inplace = True)
print(df.head())
```

The next step to cleaning our data is treating out outliers. If we sort our data by the MarketEstimate prices, you can observe the range by showcasing the first five rows and the last five rows. 

```python
# Sorting the data by market estimate values 
df.sort_values(by=['MarketEstimate'], inplace = True)
print(df.head())
print(df.tail())
```

As you can see above, the ranges for the markest estimate is very large. The range currently is 15,700 to 36,876,900. We want to do a more accurate and average prediction for our housing prices. 
According to <a href ="https://www.forbes.com/advisor/mortgages/real-estate/median-home-prices-by-state/"> Forbes </a>, the average house price in the United States was 495,100 in the second quarter of 2023, according to the Census Bureau and Department of Housing and Urban Development.

To decide what range we should change the pricing too, we'll plot a boxplot to see the distributions of prices and where the averages lay. 

```python
# Plotting the market estimate price distribution boxplot

plt.rcParams.update({'font.size': 10})

df.boxplot(column='MarketEstimate', grid=False)
plt.title('Distribution of Market Estimates')
plt.ylabel('Price')
plt.show()
```

Looking at this box plot, the small circles indicate outliers. There are a lot of high outliers above the 1,000,000 range. At first we'll try aiming our range from 100,000 to 1,000,000 range and then look at a histogram of the distribution of prices to see the finer details and frequency of the points. 

```python
# Restricting the data to market estimate prices of $100,000 to $1,000,000
df = df[(df['MarketEstimate'] >= 100000) & (df['MarketEstimate'] <= 1000000)]

# Plotting a histogram of the price distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['MarketEstimate'], bins=30, kde=True)
plt.title('Distribution of Market Estimate Prices')
plt.xlabel('Market Estimate Prices')
plt.ylabel('Frequency')
plt.show()
```

Looking at the histogram, we can justify the range of 100,000 to 1,000,000. Most of the data points are around the 500,000 range so a range from 100,000 to 1,000,000 will be fitting.

Yay! Now that we finished looking at the general trends of all our features. Now let's look at some visualizations that will help us see how our features are connected to estimating market prices. 

To see how the bathroom and bedrooms are related to our estimate park prices, we will the bathroom and bedrooms based on their counts. Then we'll calculate the average estimated house price per group to see the trend. 

```python
# Plotting the average estimated house price based on bathroom groups

# Group by 'bathrooms' and calculate average price for each group
avg_prices_by_bathrooms = df.groupby('Bathroom')['MarketEstimate'].mean().reset_index()

# Plotting the line graph
plt.figure(figsize=(15, 5)) 

# First subplot for bathroom groups
plt.subplot(1, 2, 1)  # 1 row, 2 columns, subplot 1
plt.plot(avg_prices_by_bathrooms['Bathroom'], avg_prices_by_bathrooms['MarketEstimate'], marker='o')
plt.title('Average House Price Estimate by Number of Bathrooms')
plt.xlabel('Number of Bathrooms')
plt.ylabel('Average House Price Estimate')

# Plotting the average estimated house price based on bedroom groups

# Group by 'bedrooms' and calculate average price for each group
avg_prices_by_bedrooms = df.groupby('Bedroom')['MarketEstimate'].mean().reset_index()

# Second subplot for bedroom groups
plt.subplot(1, 2, 2)  # 1 row, 2 columns, subplot 2
plt.plot(avg_prices_by_bedrooms['Bedroom'], avg_prices_by_bedrooms['MarketEstimate'], marker='o')
plt.title('Average House Price Estimate by Number of Bedrooms')
plt.xlabel('Number of Bedrooms')
plt.ylabel('Average House Price Estimate')

# Adjust layout
plt.tight_layout()

# Show the plot
plt.show()
```

We can see an increasing trend here where as the numbers of bathrooms increase, the average house price estimate increases.  We can also see an increasing trend as the number of bedrooms increase, the  average house price estimate increases. 

Now let's see the relationship between the area sizes/price per square feet and average house prices. 

```python
# This function will create ranges out of the values given and then plot the ranges against the
# average estimated house price. 
def plot_avg_price_by_range(df, column_name, xlabel):
    # Determine min and max values of the specified column
    min_val = df[column_name].min()
    max_val = df[column_name].max()

    # Define the number of bins (e.g., 10 bins)
    num_bins = 10

    # Create bins based on min and max values
    bins = np.linspace(min_val, max_val, num_bins + 1)

    # Create labels for the bins
    labels = [f'{int(bins[i])}-{int(bins[i+1])}' for i in range(len(bins) - 1)]

    # Create a new column 'range' based on the bins
    range_column = f'{column_name}_range'
    df[range_column] = pd.cut(df[column_name], bins=bins, labels=labels, include_lowest=True)

    # Group by 'range' and calculate average house price for each group
    avg_price_by_range = df.groupby(range_column)['MarketEstimate'].mean().reset_index()

    # Plotting the line graph
    plt.plot(avg_price_by_range[range_column], avg_price_by_range['MarketEstimate'], marker='o', linestyle='-')

    # Adding labels and title
    plt.title(f'Average House Price vs. {xlabel} Range')
    plt.xlabel(f'{xlabel} Range')
    plt.ylabel('Average Estimate House Price ($)')
    plt.xticks(rotation=45)  # Rotate x-axis labels for better readability


# Create a figure and two subplots side by side
fig, axes = plt.subplots(1, 2, figsize=(16, 6))  # 1 row, 2 columns

# Plotting Average House Price vs. Area Range
plt.sca(axes[0])  # Select the first subplot
plot_avg_price_by_range(df, 'Area', 'Area (sq. ft)')

# Plotting Average House Price vs. Price Per Square Feet Range
plt.sca(axes[1])  # Select the second subplot
plot_avg_price_by_range(df, 'PPSq', 'Price Per Square Feet ($)')

# Adjust layout and display the plot
plt.tight_layout()  # Adjust subplot parameters to give specified padding
plt.show()
```

As seen in our graphs above, we can see there seems to be a potential increasing linear relationship between our estimated house prices and area. This can be said the same for our price per square feet. We will see more indepth if there's a correlation in our next step of the Data Science lifecycle!
